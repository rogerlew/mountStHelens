## Read data into Notebook

We will use gdal to read the data. To do so we will create a pretty basic function to extract the data. It assumes the dataset contains a single band of data and that the single band contains float data.

```python
def readDEMdata(fname):
    ds = gdal.Open(fname, GA_ReadOnly)

    if ds is None:
        raise Exception('Could not open "%s"' % fname)

    assert ds.RasterCount == 1
    
    band = ds.GetRasterBand(1)
    return np.array(band.ReadAsArray(), dtype=np.float)
```

```python
before = readDEMdata(r"intermediates/before_downscaled.tif")
before *= 3.28084 # convert from meters to feet
before.shape, np.min(before), np.max(before) # check to make sure there is data
```

```python
after = readDEMdata(r"mtsthelens_after/10.2.1.1043901.dem")
after.shape, np.min(after), np.max(after)
```

## Crop the datasets

Let's crop the image to be centered about the peak of the volcano

```python
# need to identify the peak
# argmax returns the index of the maximum value, then unravel_index finds the 2d coordinates for us
np.unravel_index(before.argmax(), before.shape)
```

```python

# I want to crop about the peak location. The value determined above looks reasonable
x0, y0 = 575, 467
n = 300 # Let's go 600 pixels (6,000 m) from this center point

# use array slices to perform crop
before = before[x0-n:x0+n, y0-n:y0+n]
after = after[x0-n:x0+n, y0-n:y0+n]

# check that there aren't any Nodata values present
np.min(before), np.max(before), np.min(after), np.max(after)
```

## Colorize the datasets

One of the goals of this exercise is to compare matplotlib's default Rainbow colormap to colormaps that take into account
perception of color and luminance. To do thise we want to create the set of plots that are exactly the same except for the
colormap that is applied to the data.

```python
def plotSaintHelensData(cmap="CubicYF", outdir="diptychs", vmin=3800, vmax=10000, 
                        colorbar_ticks=np.linspace(4000,10000,7)):
    
    # make it clear were these global values come from
    global before, after
    
    # find the colormap. The pmkmp module registers new colormaps with matplotlib.
    cm = plt.get_cmap(cmap)

    # declar a new figure
    plt.figure(figsize=(16,8))
    
    # set the subplot margins
    plt.subplots_adjust(bottom=0.01, right=0.82, top=0.95, wspace=0.08)

    # create side by side subplots
    plt.subplot(121)
    plt.imshow(before, cmap=cm, vmin=vmin, vmax=vmax)
    plt.xticks([0, 150, 300, 450, 600], ["-300", "-150", "0", "150", "300"])
    plt.xlabel('meters')
    plt.yticks([0, 150, 300, 450, 600], ["-300", "-150", "0", "150", "300"])
    plt.ylabel('meters')
    plt.title("Before 5/18/80")

    plt.subplot(122)
    plt.imshow(after, cmap=cm, vmin=vmin, vmax=vmax)
    plt.xticks([0, 150, 300, 450, 600], ["-300", "-150", "0", "150", "300"])
    plt.xlabel('meters')
    plt.title("After 5/18/80")
    plt.yticks([])
    plt.text(-30, -70, "Mount Saint Helens", horizontalalignment="center", fontsize=15)

    # create the colorbar
    cax = plt.axes([0.85, 0.12, 0.03, 0.72])
    cbar = plt.colorbar(cax=cax, cmap=cm, ticks=colorbar_ticks)
    
    # save the results
    if not os.path.exists(outdir):
        os.mkdir(outdir)
    plt.savefig(os.path.join(outdir, 'mtStHelens_%s.png' % cmap), dpi=300)
```

### CubicYF

Up first is the pmkmp.CubicYF colormap. The author of the colormap has a nice essay on how it was created here:

https://mycarta.wordpress.com/2013/02/21/perceptual-rainbow-palette-the-method/

I've simply made migrated his matlab code to matplotlib.

The luminance of the colormap is monotonic and roughly follows a inverse cubic function.

This has become my new default.

```python
plotSaintHelensData()
```

### jet

Now let's compare this to the jet color map. Here we can see the luminance of the cyan is much greater than any other part of the colormap. It makes the mid-elevation values much more salient and sucks out the details in this region.

```python
plotSaintHelensData("jet")
```

### LinearL

And finally we have the pmkmp.LinearL colormap. as the name implies it has a linear luminance ramp over just about the entire range of values representable by RGB space. 

```python
plotSaintHelensData("LinearL", vmax=10500) # highlights on peak were blown so the vmax was adjusted slightly
```

### cubehelix

Here is another colormap with linear luminance. This one is part of matplotlib and might already be on your machine.

## Encore

Often times baking light shading into an image can help to accentuate details that would otherwise be missed. This little class extends matplotlib.colors.LightSource for producing shaded colorized hillshades

```python
from matplotlib.colors import LightSource, Normalize, hsv_to_rgb, rgb_to_hsv

class LightSourceExt(LightSource):
    def __init__(self, azdeg=315, altdeg=45,
                 hsv_min_val=0, hsv_max_val=1, hsv_min_sat=1,
                 hsv_max_sat=0):

        """
        subclass of matplotlib.colors.LightSource
        
        Specify the azimuth (measured clockwise from south) and altitude
        (measured up from the plane of the surface) of the light source
        in degrees.

        The color of the resulting image will be darkened
        by moving the (s,v) values (in hsv colorspace) toward
        (hsv_min_sat, hsv_min_val) in the shaded regions, or
        lightened by sliding (s,v) toward
        (hsv_max_sat hsv_max_val) in regions that are illuminated.
        The default extremes are chose so that completely shaded points
        are nearly black (s = 1, v = 0) and completely illuminated points
        are nearly white (s = 0, v = 1).
        """
        
        LightSource.__init__(self, azdeg=azdeg, altdeg=altdeg,
                             hsv_min_val=hsv_min_val, hsv_max_val=hsv_max_val, 
                             hsv_min_sat=hsv_min_sat, hsv_max_sat=hsv_max_sat)
        
    def shade_elevation(self, elevation, cm, vmin, vmax, fraction=1.0):
        """
        Take the input elevation array (ny*nx) and adjust values
        to given the impression of a shaded relief map with a
        specified light source using the elevation (ny*nx).
        A new value array is returned.
        """

        # imagine an artificial sun placed at infinity in some azimuth and
        # elevation position illuminating our surface. The parts of the
        # surface that slope toward the sun should brighten while those sides
        # facing away should become darker. convert alt, az to radians
        az = self.azdeg * np.pi / 180.0
        alt = self.altdeg * np.pi / 180.0
        # gradient in x and y directions
        dx, dy = np.gradient(-elevation)
        slope = 0.5 * np.pi - np.arctan(np.hypot(dx, dy))
        aspect = np.arctan2(dx, dy)
        intensity = (np.sin(alt) * np.sin(slope) + np.cos(alt) *
                     np.cos(slope) * np.cos(-az - aspect - 0.5 * np.pi))
        # rescale to interval -1,1
        # +1 means maximum sun exposure and -1 means complete shade.
        intensity = (intensity - intensity.min()) / \
                    (intensity.max() - intensity.min())
        intensity = (2. * intensity - 1.) * fraction
        
     
        norm = Normalize(vmin, vmax)
        
        rgb = cm(norm(elevation))
        
        hsv = rgb_to_hsv(rgb[:,:,:3])
        
        v = np.zeros(elevation.shape)
        # modify hsv values to simulate illumination.

        hsv[:, :, 1] = np.where(np.logical_and(np.abs(hsv[:, :, 1]) > 1.e-10,
                                               intensity > 0),
                                ((1. - intensity) * hsv[:, :, 1] +
                                 intensity * self.hsv_max_sat),
                                hsv[:, :, 1])

        hsv[:, :, 2] = np.where(intensity > 0,
                                ((1. - intensity) * hsv[:, :, 2] +
                                 intensity * self.hsv_max_val),
                                hsv[:, :, 2])

        hsv[:, :, 1] = np.where(np.logical_and(np.abs(hsv[:, :, 1]) > 1.e-10,
                                               intensity < 0),
                                ((1. + intensity) * hsv[:, :, 1] -
                                 intensity * self.hsv_min_sat),
                                hsv[:, :, 1])
        hsv[:, :, 2] = np.where(intensity < 0,
                                ((1. + intensity) * hsv[:, :, 2] -
                                 intensity * self.hsv_min_val),
                                hsv[:, :, 2])
        hsv[:, :, 1:] = np.where(hsv[:, :, 1:] < 0., 0, hsv[:, :, 1:])
        hsv[:, :, 1:] = np.where(hsv[:, :, 1:] > 1., 1, hsv[:, :, 1:])
        # convert modified hsv back to rgb.
        return hsv_to_rgb(hsv)
```

```python
def plotShadedSaintHelensData(cmap="CubicYF", outdir="diptychs", vmin=3800, vmax=10000, az=315, fraction=0.7):
    
    # make it clear were these global values come from
    global before_blurred, after
    
    ls = LightSourceExt(azdeg=az)
    
    # find the colormap. The pmkmp module registers new colormaps with matplotlib.
    cm = plt.get_cmap(cmap)

    # declar a new figure
    plt.figure(figsize=(16,8))
    plt.set_cmap(cm)
    
    # set the subplot margins
    plt.subplots_adjust(bottom=0.01, right=0.82, top=0.95, wspace=0.08)

    # create side by side subplots
    plt.subplot(121)
    rgb = ls.shade_elevation(before_blurred, cm, vmin, vmax, fraction=fraction)
    plt.imshow(rgb)
    plt.xticks([0, 150, 300, 450, 600], ["-300", "-150", "0", "150", "300"])
    plt.xlabel('meters')
    plt.yticks([0, 150, 300, 450, 600], ["-300", "-150", "0", "150", "300"])
    plt.ylabel('meters')
    plt.title("Before 5/18/80")

    plt.subplot(122)
    rgb = ls.shade_elevation(after, cm, vmin, vmax, fraction=fraction)
    plt.imshow(rgb)
    plt.xticks([0, 150, 300, 450, 600], ["-300", "-150", "0", "150", "300"])
    plt.xlabel('meters')
    plt.title("After 5/18/80")
    plt.yticks([])
    plt.text(-30, -70, "Mount Saint Helens", horizontalalignment="center", fontsize=15)

    # create the colorbar
    cax = plt.axes([0.85, 0.12, 0.03, 0.72])
    cbar = plt.colorbar(cax=cax, cmap=cm, ticks=np.linspace(0.01,0.995,7))
    cbar.ax.set_yticklabels([str(int(x)) for x in np.linspace(4000,10000,7)])# vertically oriented colorbar
  
    
    # save the results
    if not os.path.exists(outdir):
        os.mkdir(outdir)
    plt.savefig(os.path.join(outdir, 'mtStHelens_Shaded%s.png' % cmap), dpi=300)
```
